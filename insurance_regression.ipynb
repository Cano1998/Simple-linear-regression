```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Import data
ins = pd.read_csv('insurance.csv')
ins.head()
```

```python
# Select the columns 'age' and 'charges' for plotting
x = ins[['age']].values
y = ins[['charges']].values

plt.scatter(x,y,color = 'y')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Fit the model and plot the data and model
charges_model = LinearRegression()
charges_model.fit(x,y)
y_pred = charges_model.predict(x)

plt.scatter(x,y,color = 'y')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Predict an unknown value (maximum age in data set is 64) 
unk_x = [[70]]

x_pred = np.append(x, unk_x).reshape(-1, 1)
y_pred = charges_model.predict(x_pred)

plt.scatter(x,y,color = 'y')
plt.plot(x_pred, y_pred, color='r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

print('The charge for a 70 years old person should be: ', charges_model.predict(unk_x))
```
